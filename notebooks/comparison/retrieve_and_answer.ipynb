```python
import os
import sys
from pprint import pprint 

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '../..')))

from app_settings import PROJECT_ROOT
from src.g_retriever import retrieve

os.chdir(PROJECT_ROOT)
print(f"Changed working dir to {PROJECT_ROOT}")
```

```python
questions = ["What are some of the projects they work at Granter?",
             "Who leads the client team at Granter?",
             "What technologies do they use at Granter?"]
```

```python
retrieved = retrieve(
        graph_csv='results/KGs/rdb_ontology_kg.csv',
        question=questions[0],
        embed_model='paraphrase-multilingual-MiniLM-L12-v2',
        topk=10
    )

print(retrieved)
```

```python
from langchain_openai import AzureChatOpenAI
from app_settings import (
    AZURE_OPENAI_API_KEY,
    AZURE_OPENAI_ENDPOINT,
    AZURE_OPENAI_API_VERSION,
    AZURE_DEPLOYMENT_GPT41_NANO
)

gpt41_nano = AzureChatOpenAI(
    azure_deployment=AZURE_DEPLOYMENT_GPT41_NANO,
    api_version=AZURE_OPENAI_API_VERSION,
    azure_endpoint=AZURE_OPENAI_ENDPOINT,
    api_key=AZURE_OPENAI_API_KEY
)
```

```python
prompt = f"Question: {questions[0]}\n\nContext:\n{retrieved}\nAnswer:"

response = gpt41_nano.invoke(prompt)
pprint(response.content)
```

```python
for question in questions:
    retrieved = retrieve(
        graph_csv='results/KGs/txt_ontology_kg_nonstrict.csv',
        question=question,
        embed_model='paraphrase-multilingual-MiniLM-L12-v2'
    )
    print(retrieved)
    prompt = f"Question: {question}\n\nContext:\n{retrieved}\nAnswer:"
    response = gpt41_nano.invoke(prompt).content
    print('\n\nRESPONSE:\n\n')
    pprint(response)
```

```python
import io

output_path = "retrieval_graph_withchunks.txt"
with io.open(output_path, "w", encoding="utf-8") as f:
    for question in questions:
        retrieved = retrieve(
            graph_csv='results/KGs/rdb_ontology_kg.csv',
            question=question,
            embed_model='paraphrase-multilingual-MiniLM-L12-v2'
        )
        f.write(f"Question: {question}\n")
        f.write(f"Context:\n{retrieved}\n")
        prompt = f"Question: {question}\n\nContext:\n{retrieved}\nAnswer:"
        response = gpt41_nano.invoke(prompt).content
        f.write("\nRESPONSE:\n\n")
        f.write(f"{response}\n")
        f.write("="*40 + "\n")
```
